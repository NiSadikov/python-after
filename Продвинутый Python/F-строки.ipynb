# Задача 1 

Оказалось, что в выпускном классе школы города X девять учеников закончили
школу с отличием. Произошло невероятное совпадение, что у выпускников с красными дипломами на всех три имени, три фамилии и одно отчество:

Помогите администрации школы не запутаться при выдаче аттестатов, написав
программу, которая выводит на экран строку в следующем формате для каждого
выпускника:
"Диплом с отличием вручается Вотякову Александру Романовичу."
Примечание. Не забывайте про склонение имён.

```python
name = ['Александру','Алексу','Альберту']
surname = ['Вотяку','Вотякову','Вотяковичу']
patronymic = 'Романовичу'
for x in surname:
    for y in name: 
        print(f'Диплом с отличием вручается {x} {y} {patronymic}')
```

# Задача 2

имеют следующий формат, где X соответствует буквам латинского алфавита, а 0
соответствует десятичным цифрам:

"XXX0000-000"

Напишите программу, которая принимает на вход три значения A, B и C, каждое
на отдельной строке, а затем выводит соответствующий этим значениям уникальный
номер. Значение А – первые три буквы номера в виде строки, B – неотрицательное
число, соответствующее первым четырём цифрам номера и С – неотрицательное
число, соответствующее последним трём цифрами номера.

```python
print(f'{input()}{int(input()):04}-{int(input()):03}')
```

# Задача 3 

Саша в этом году переходит в 4-й класс и впервые столкнулся с языком Python. Он
узнал, что с помощью f-строк можно не только автоматически выравнивать числа
по правому краю при сложении их в столбик, но и сразу же получать правильный
ответ.
Помогите Саше написать программу, которая бы брала на вход два числа
(A, B < $10^{9}$) на отдельных строках и выводила в консоль процесс эти два числа,

# Задача 4 

Артём 1-го января 2023-го года положил на счёт в банке 100 000 000 рублей под
r%, который выплачивается в конце каждого месяца. Напишите программу, которая
считает, сколько денег будет на вкладе через k месяцев. На вход подаются два неотрицательных числа – r и k, где r – процент выплат каждый месяц, а k – количество
месяцев, которое прошло с открытия вклад

```python
s = 100000000
r = int(input())
k = int(input())
print(f'{s*((1+(r/100))**k):,.2f}')
```

Два друга, Петя и Ваня, решили написать функцию, которая вычисляет всю таблицу умножения от 1 до 100. Чтобы удостовериться, что всё работает правильно, они
хотят вывести на экран дебаг-строки для всех результатов произведений, в которых
есть цифра 0.
Помогите Пете и Ване написать программу и проверить её на правильность, добавив дебаг-строки в следующем формате:
"[DEBUG] a=1 b=10 result=10"

```python
for a in range(1,101):
    for b in range(1,101):
        result = a*b
        if '0' in str(result):
            print(f"[DEBUG] {a=} {b=} {result=}")
    
```

 # Задача 6 
 
 После того, как Петя и Ваня потратили 4 дня, чтобы разобраться с задачками на
основную теорему арифметики, они решили вычислить создателя этого прототипа
по IP.
Создатель этих задачек предвидел такое развитие событий и установил защиту от
вычисления по IP, высылая вместо этого IP серверов из Майнкрафта. Он знает, что
ни Петя, ни Ваня, не знают, как выглядят IP, поэтому запутать их будет несложно.
Напишите функцию, которая принимает на вход 4 числа a, b, c и d (0 ≤ a, b, c, d <
$2^{8}$
) на одной строке, а затем возвращает IP в следующих форматах, каждое на новой
строке:
• Восьмибитном двоичном: 01111111.00000000.00000000.00000001
• Двоичном: 1111111.0.0.1
• Восьмиричном: 177.0.0.1
• Десятичном: 127.0.0.1
• Шестнадцатиричном: 7f.0.0.1

Примечание. Используйте метод .format()

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())
print('{:08b}.{:08b}.{:08b}.{:08b}'.format(a,b,c,d))
print('{:b}.{:b}.{:b}.{:b}'.format(a,b,c,d))
print('{:o}.{:o}.{:o}.{:o}'.format(a,b,c,d))
print('{}.{}.{}.{}'.format(a,b,c,d))
print('{:x}.{:x}.{:x}.{:x}'.format(a,b,c,d))
```
