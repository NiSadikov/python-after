Пётр Паролевич в очередной раз забыл свой пароль. К счастью, он нашёл листочек, на котором записывает все свои пароли. Каждый пароль может состоять из
латинских строчных букв и десятичных цифр.
Петр Паролевич помнит, что его пароль – шестнадцатеричное число. По рассеянности Пётр Паролевич начал вводить все пароли подряд. Помогите Петру разобраться, какие строки могут являться его паролями.
Напишите функцию, которая принимает в качестве аргументов сколько угодно
строк, и возвращает список со всеми значениями, которые могут быть паролями
Петра Паролевича.
Примечание. Для проверки пароля можно использовать int(password, 16).
Если число не подходит, тогда нужно обработать исключение ValueError.

```python
def password_verification(*passwords):
    array = []
    for password in passwords: 
        try: 
            int(password, 16)
            array.append(password)
        except: 
            pass
    return array
print(password_verification("password", "123456", "abcdef", "bcdefg"))
```

```python
olympiad1 = {"name": "Пробная вышка",
"winners": {
"Олеся Олимпиадникова": 594,
"Олег Олимпиадников": 587,
"Онисим Олимпиадников": 581,
}
}

olympiad2 = {"name": "Горные воробьи",
"winners": {
"Ольга Олимпиадникова": (20, 20, 19, 20),
"Олеся Олимпиадникова": (19, 19, 20, 20, 17),
"Офелия Олимпиадникова": (20, 20, 20, 20, 13)
}
}

def checking_the_results(*dictionaries):
    try: 
        print(f"Результаты Ольги Олимпиадниковой {dictionaries[0]['name']} : {dictionaries[0]['winners']['Ольга Олимпиадникова']}")
    except KeyError: 
        print('Ольга Олимпиадникова не выиграла в Пробной вышке, но стала призером')
    try: 
        print(f"Баллы за 5 задачу Ольги Олимпиадниковой в Горных воробьях: {dictionaries[1]['name']} : {dictionaries[1]['winners']['Ольга Олимпиадникова'][4]}")
    except KeyError: 
        print('Ольга Олимпиадникова не выиграла в Горных воробьях')
    except IndexError: 
        print('Результаты Ольги Олимпиадниковой за 5-ю задачу в Горных воробьях не опубликовали, но она выиграла в олимпиаде')        
    try: 
        print(f"Результаты Олеси Олимпиадниковой {dictionaries[0]['name']} : {dictionaries[0]['winners']['Олеся Олимпиадникова']}")
    except KeyError: 
        pass
    try: 
        print(f"Баллы за 5-ю задачу Олеси Олимпиадниковой в {dictionaries[1]['name']} : {dictionaries[1]['winners']['Олеся Олимпиадникова'][4]}")
    except (KeyError, IndexError): 
        pass
    
checking_the_results(olympiad1,olympiad2)
```

# Задача 5 

Реализуйте программу, которая соответствует функционалу известного анекдота.
У этого задания нет единственного способа решения. Вы можете сами выбрать, какие
исключения возвращать и что делать в каждом из случаев. Можно использовать
комментарии под строками анекдота или самостоятельно придумать чтонибудь уникальное.
Что рекомендуется использовать:

1. инструкцию raise;
2. инструкцию assert, возвращая правильные сообщения об ошибке;
3. классы типа Exception для обозначения уникальных ошибок;
4. инструкцию finally для вывода сообщения о результате действия.

Анекдот.

Заходит однажды тестировщик в бар.
Заказывает:
кружку лимонада,
2 кружки лимонада,
0 кружек лимонада,
999999999 кружек лимонада,
ящерицу в стакане,
–1 кружку лимонада,
qwerty кружек лимонада.
Первый реальный клиент заходит в бар и спрашивает, где туалет. Бар вспыхивает
пламенем

```python
class LizardInCup(Exception):
    pass
class BarBurntDown(Exception):
    pass
def orderBeer():
    try:
        mugs = input("\nВы заходите в бар. \
        Сколько кружек лимонада заказываете?\n").lower()
        if mugs == "ящерица в стакане":
            raise LizardInCup
        if mugs == "где туалет?":
            raise BarBurntDown
        mugs = int(mugs)
        assert mugs < 2, "Кризис. Не больше одной кружки в одни руки."
        if mugs <= 0 or mugs >= 100:
            raise ValueError
    except AssertionError:
        print("Кризис. Не больше одной кружки в одни руки.")
    except ValueError:
        print("Вы не можете заказать такое число кружек, \
    держите одну кружку.")
    except LizardInCup:
        print("У нас закончились ящерицы. Приходите завтра!")
    except BarBurntDown:
        print("Поздравляем! Бар сгорел. Вы вышли из симуляции.")
        return
    print("Приходите ещё!")
    orderBeer()
orderBeer()
```
