```python
def whole_file(name): #1
    with open(name,'r',encoding='utf-8') as f:
        return f.read()

def first_line(name): #2
    with open(name,'r',encoding='utf-8') as f:
        return f.readline()

def list_string(name): #3
    with open(name,'r',encoding='utf-8') as f:
        return f.readlines()

def list_string_withoutHyphenation(name): #4
    with open(name,'r',encoding='utf-8') as f:
        return f.read().split('\n')

def line_by_lineOutput(name): #5
    with open(name,'r',encoding='utf-8') as f:
        for x in f:
            print(x, end='')
        
def connect_with_Spaces(name): #6
    with open(name,'r',encoding='utf-8') as f:
        k = f.read().split('\n')
        return ' '.join(k)
    
def perfect_string(name): #7
    return name.rstrip('n')

def without_emotions(name): #8
    return name.rstrip('!?.')

def enter_line(name): #9
    with open('def.txt','w',encoding='utf-8') as f:
        return f.write(name)
    
def enter_line_hyphen(name): #10
    with open('def2.txt','w',encoding='utf-8') as f:
        return f.write(name+'\n')
        
def enter_line_everything(name): #11
    with open('def3.txt','w',encoding='utf-8') as f:
        return f.writelines(name)

def two_files(one,two): #12
    with open(one,'r',encoding='utf-8') as f:
        with open(two,'w',encoding='utf-8') as g:
            for line in f: 
                print(line.rstrip(),file=g)

def two_files_end_start(one,two): #13
    with open(one,'r',encoding='utf-8') as f:
        with open(two,'w',encoding='utf-8') as g:
            for line in f:
                if line.startswith('hello') and line.endswith('world'):
                    print(line.strip(),file=g)                
            
def dictionary(name): #14
    with open(name, 'r', encoding='utf-8') as f:
        dictir = {}
        k = f.readlines()
        for x in k[1:]:
            s = x.rstrip().split() 
            dictir[s[0]] = (s[1], int(s[2]))
        return dictir
```
