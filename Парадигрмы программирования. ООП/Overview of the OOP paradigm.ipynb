# Обзор парадигмы ООП

Говорят, чтобы проверить, поняли ли вы тему хорошо или нет, нужно попытаться её объяснить другим. Для полного понимания парадигмы ООП, вас попросили привести примеры к некоторым основным концепциям. Попробуйте притвориться Александром Романовичем и придумать примеры, которые были бы понятны школьникам средних классов.



## Задача 1

Попытайтесь сформулировать объяснения максимально просто: проводите паралелли с объектами реального мира и действиями, которые они выполняют. Вам нужно описать следующие базовые концепции:

1) класс;  
2) параметры класса;  
3) функции класса  

Давайте порассуждаем о следующих концепциях: класс, параметры класса, функция класса. Понятие класса можно характеризовать следующим образом: когда мы хотим построить автомобиль, то нам необходим его чертеж, который будет отражать основные концепции будущих автомобилей, таким чертежем выступает понятие класса: в нём зафиксирована основная функциональнсть каждого автомобиля, что будет сделан с помощью этого чертежа. Параметрами класса выступают инициированные свойства, которыми обладают автомобили: у них 4 колеса, руль, лобовое стекло. Помимо этого, у класса есть функции, которые отвечают за взаимодействием с автомобилем: например, функция открытия двери или функция открытия багажника. 

## Задача 2

Придумайте прикольные паралелли между предметами, которые можно найти и использовать в школе, чтобы объяснить понятие инкапсуляции для учеников, которые хотят всерьёз заниматься программированием.

Вернёмся к нашим автомобилям, когда мы ими пользуемся, мы по сути выполняем всего несколько действий: жмём на педали, крутим руль, совершенно не задумываясь о тысяче операций, скрытых от нашего взора: этот принцип лежит в основе концепции инкапсуляции, что является возможность скрыть из интерфейса пользователя внутренние функции, которые используются для поддержания работы системы, но не участвуют в непосредственном взаимодействии с человеком: мы не видим провода, соеднинения автомобиля, не знаем, как работает двигатель, но всё равно управляем им через видимые нам функции: руль, педали. 

## Задача 3 

Оказалось, что большая часть школьников, которым вам нужно объяснить, как работает наследование в ООП, являются визуалами. Возьмите какую-нибудь группу объектов и изобразите иерархию объектов в виде графа, схемы или дерева, которое бы визуализировало наследование. Для каждого описанного объекта и подобъекта стоит указать следующее:

1) название класса;  
2) параметры, которые являются общими у всех подклассов каждого класса;  
3) параметры, которые различаются в зависимости от класса одной подгруппы;  
4) проведите параллели между объектами в реальном мире и описанными абстрактными классами.  

Мы можем рассмотреть класс автомобилей и его подклассы: легковые автомобили, грузовые автомобили и т.д. Концепция наследования заключается в том, что подклассы наследуют основную функциональность своего родительского класса: у всех есть руль, у всех есть колёса, но тем не менее все подклассы различаются между собой: например, у каждого автомобиля свой цвет, свой объем топливного бака, разная яркость фар.

# Задача 4

После прекрасно проведённого занятия некоторые ученики из группы, которая работала над визуализацией иерархии животных в природе, остались после занятия и завалили вас вопросами:

что делать, если разных животных нужно кормить по-разному?
вот бывает же такое, что хозяева питомцев по-разному воспитывают их, а как это учесть?
помогите пожалуйста, а что делать, если у меня получается, что все животные ходят на четырёх лапах, как это исправить?

Вам стало очень приятно, что ученики задают такие умные вопросы, поэтому вы решили, что темой следующего урока будет какая-то определённая парадигма программирования, которая ответит на все эти вопросы.

Определите, что это за парадигма и объясните детям её базовые концепты.

Парадигма, о которой идёт речь - объективно ориентированная парадигма программирования, которая включает в себя классы, функции классов, параметры классов и такие важные концепции, как полиморфизм, наследование и инкапсуляция. Полиморфизм позволяет определённую функцию расширить на все подклассы, но при этом, сделав функцию индивидуальной для каждого подкласса. Например, всех животных, которые входят в широкий класс animal, необходимо кормить, но каждому - кормить разным кормом, в основе этого лежит полиморфизм: возможность индивидуализации родительской функции для каждого подкласса. Базовая концепция объектно-ориентированного программировании лежит в переложении взаимодействия объектов друг с другом на язык функций, параметров и классов.
